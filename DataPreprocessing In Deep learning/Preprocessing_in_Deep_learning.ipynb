## Normalization via Mini-Batch Statistic

### Two necessary simplifications:
#### 1- Normalizaing each scalar feature independently.
#### 2- Using mini-batches in stochastics gradient training.

 The meain advantage of batch-normalization is by using it we can change the parameters of network with no fear of destroying the neural network result. For example with batch-normalization we are able to use higher learning rate. and keep weights in a good distribution.

 It most of the time use after fully connected layer or convolutional layer and befor nonlinearity.

#### One disadvatage of batch-normalization:
 Always try to limit data in a particular domain so make the neural network a bit weaker.

tanh() and sigmoid() are two popular activation functions but they have a problem It called saturating,gradient going to zero so there is no update.


 Batch-Normalization can hold the data in a good bound for sigmoid() and tanh() to avoid saturation so it solve the saturation problem.

## In Conclude:
